```python
df = df.drop(['PassengerId', 'Ticket', 'Cabin'], axis = 1)
```

```python
df.isnull().sum()
```

```python
df.Age.fillna(df.Age.median(), inplace = True)
```

```python
df.Fare.fillna(df.Fare.median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.Embarked.value_counts()
```

```python
df.Embarked.fillna(df.Embarked.mode()[0], inplace = True)
df.isnull().sum()
```

```python
df['FamilySize'] = df.SibSp + df.Parch

df['IsAlone'] = 1 
df['IsAlone'].loc[df['FamilySize'] > 1] = 0 

df['Title'] = df['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]

df['AgeBin'] = pd.cut(df['Age'].astype(int), 5)
df['FareBin'] = pd.qcut(df['Fare'], 4)

df.head()
```

```python
df.Title.value_counts()
```

```python
title_names = (df.Title.value_counts() < 10)

df['Title'] = df['Title'].apply(lambda x: 'Other' if title_names.loc[x] == True else x)
print(df['Title'].value_counts())
```

```python
df.drop('Name', axis = 1, inplace= True)
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics
```

```python
target = ['Survived']
```

```python
label = LabelEncoder()
df['Sex'] = label.fit_transform(df['Sex'])
df['Embarked'] = label.fit_transform(df['Embarked'])
df['Title'] = label.fit_transform(df['Title'])
df['Age'] = label.fit_transform(df['AgeBin'])
df['Fare'] = label.fit_transform(df['FareBin'])

df.head()
```

```python
df.drop(['FareBin', 'AgeBin'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
columns_train = [col for col in df.columns.tolist() if col not in target]
columns_train
```

```python
df = pd.get_dummies(df, columns = columns_train)
df.head()
```

## Modelling

```python
label = df["Survived"]
train = df.drop(labels = ["Survived"], axis = 1)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(train, label, test_size = 0.20, random_state = 13) 
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear') # Linear Kernel

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# with grid search
kfold = model_selection.StratifiedKFold(n_splits=10)


SVMC = svm.SVC(probability=True)
svc_param_grid = {'kernel': ['linear', 'rbf'], 
                  'gamma': [ 0.001, 0.01, 0.1, 1],
                  'C': [1, 10, 50, 100,200,300, 1000]}

gsSVMC = model_selection.GridSearchCV(SVMC, param_grid = svc_param_grid, cv = kfold, scoring="accuracy", n_jobs= 2, verbose = 4)

gsSVMC.fit(X_train, y_train)

SVMC_best = gsSVMC.best_estimator_

# Best score
gsSVMC.best_score_
```
